# Дослідження динаміки теплопровідності в електронних системах за допомогою графів
## 1. Постановка задачі
У цьому ми моделюємо розподіл тепла на комп'ютерній платі (або будь-якій системі, яку можна дискретно зобразити). Замість неперервних матеріалів та суцільних об'єктів, ми розглядаємо дискретну, трішки полегшену модель задачі - мережевий граф.
**Вершини графа** - будуть нашими фізичними компонентами системи, які можуть нагріватися або навпаки охол джуватись. 
**Ребра графа** - це з'єднання між компонентами, умовно теплотрубки або мідні проводи. Як раз ребра визначатимуть як швидко буде змінюватися температура компонент.
## 2. Математична модель
### 2.1 Випадок для конкретної вершини
Нехай маємо $n$-ну кількість вершин, які будемо позначити як $u_i$. Температура кожної компоненти в момент часу $t$ буде описуватись як функція від $t$:
$$u_i(t)$$
Оскільки температура зміннюється з часом, ці зміни описуються похідними. Отже, ми дослідимо $\frac{du_i(t)}{dt}$ - **похідна функції** $u_i(t)$, яка показує як швидко змінюється температура вершини $i$. Логічно випливає зміст:
* Якщо $\frac{du_i(t)}{dt} > 0$, вершина нагрівається;
* Якщо $\frac{du_i(t)}{dt} < 0$, вершина охолуджється;
* Якщо $\frac{du_i(t)}{dt} = 0$, температура в цей момент часу не змінюється.

Але, окрім цього, ми кажемо:
* Є вершини $u_j$ та $u_i$. Якщо між ними є ребро, то температура одного впливає на іншого.
* Якщо $u_j > u_i$, тобто температура $j$-того більша за $i$-того, то вершина $i$ буде отримувати тепло від $j$;
* Якщо $u_j < u_i$, тоді навпаки $i$ віддає тепло $j$.
Відповідно, зміна температури буде визначатися від різниці температур двох вершин:
$$u_j - u_i$$
Де якщо:
* $u_j - u_i > 0$, то $j$ гарячіший, тепло іде до $i$
* $u_j - u_i < 0$, то аналогічно навпаки.

Тепер, наша модель набуває такого вигляду:
$$\frac{du_i}{dt} = \sum_{j \sim i} (u_j - u_i)$$,
де $j \sim i$ означає що $j$ сусід $i$. Сума в нас береться, оскільки нам потрібні відношення температур всіх сусідів до нашого $i$-тої компоненти.
### 2.2 Узагальнене рівняння на всі вершини
Для простішого моделювання, нам всі наші рівняння буде зручно описати матрично. Тому, спочатку введемо вектор-функцію, яка буде описувати одразу всі залежності температури вершин від часу:
$$u(t)= \begin{pmatrix} u_1(t) \\ u_2(t) \\ ... \\ u_n(t) \end{pmatrix}$$
Далі, зобразимо суму по сусідам як матрицю, для цього введемо спочатку матрицю суміжності графа та матрицю степенів. Матриця суміжності $A$ - це матриця яка показує які вузли з'єднанні:
$$A_{ij} = \begin{cases} 1, \ \text{якщо існує ребро між} \ i \ \text{та} \ j \\ 0, \ \text{інакше} \end{cases}$$ 
Та діагольна матриця степенів - це матриця, на діагоналі якої стоять степені вершин, числа які показують скільки сусідів має кожна вершина.

Матриця Лапласа $L = D - A$ - це наш центральний об'єкт. Ця матриця, це спосіб записати скільки вузол віддає температури через свої зв'язки та скільки, відповідно, отримує від сусідів.
Тепер, можна всі наші $n$ рівнянь записати легко, ось так:
$$\frac{du(t)}{dt} = -Lu(t)$$
